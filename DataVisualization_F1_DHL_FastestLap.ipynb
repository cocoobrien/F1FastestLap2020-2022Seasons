## Distribution of Fastest Lap Times over 2020-2022 Seasons

```python
sns.set(style='dark')
sns.histplot(df['Lap Time'], kde=True)
plt.ylabel("Count of Lap Time Occurences 2020-2022")
plt.xlabel("Lap Time (mm.ss)")
plt.title("Fastest Lap Time Distribution, 2020-2022 Seasons")
plt.show
```

The distribution above shows that most drivers obtain lap times between 1.1 and 1.3. Out of a dataset with 60 lap times, there were 15 recorded laps at about 1 minute and 20 seconds. The greatest number of laps were recorded with times betwen 1.1 and 1.3; however, this still seems to show a unimodal distribution while skewed left. There is a clear outlier around 0.6, causing its left skew, and showing that this driver's time was extremely fast achieveing a record time that occured once thorughout the last three seasons of Formula One racing. 

## Fastest Laps by Car Racing Team

```python
sns.set(style='dark')
sns.scatterplot(x="Lap Time", y="Car", data = df, hue = "Year")
plt.ylabel("Car")
plt.xlabel("Lap Time (mm.ss)")
plt.title("Fastest Formula 1 Lap Times by Car Racing Team (2020-2022)")
plt.show
```

It is clear here that there are two extremely dominant racing teams in the Formula One world for the last couple of years: Mercedes and Red Bull Racing. If one is familiar with the makeup of these teams, the last three seasons have shown close contention and rivalry between top drivers, Max Verstappen and Lewis Hamilton - who some record as the most talented Formula One drivers in the history of the sport. 

Ferrari and McLaren follow suit as stealthy competitors and racing teams to be aware of in terms of speedy performance and steady presence in the top of the grid. 

## Fastest Laps by Driver

```python
sns.scatterplot(x="Lap Time", y="Driver", data = df, hue = "Year")
plt.ylabel("Driver")
plt.xlabel("Lap Time (mm.ss)")
plt.title("Fastest Formula 1 Lap Times by Driver (2020-2022)")
plt.show
```

From this visualization, it's interesting to note the emergence of new drivers and their skill levels in speed over the past 3 seasons. Lewis Hamilton, an icon in the F1 world, clearly dominated in the 2020 and 2021 seasons. However, from the differing colors to emulate the years, you may notice another - perhaps even more - dominant figure, Max Verstappen. While he has fewer "pink" or 2020 dots on the plot, he shows strong performances in the 2021 and 2022 seasons in terms of speed. He was the winner of 'DHL Fastest Lap' with over 4 fastest laps recorded in the 2022 season, while Lewis Hamilton was the winner of 2020 and 2021. An interesting rivalry between these two to definitely keep an eye on this 2023 season. Further, two pretty young drivers seem to be on the rise: Charles LeClerc and George Russell, only counting fast laps in the 2022 seasons. Similarly, Zhou Guanyu and Carlos Sainz seemed to do well in the 2022 season. These are drivers to the on the look out for this 2023 season.

## Top 20 Lap Times from 2020-2022 Seasons

```python
# Sorting dataframe by Lap Time, shortest-longest
top_times = df.sort_values(by=['Lap Time'])
top_20_times = top_times[:20]
data = top_20_times
data
```

It is interesting to note from the chart of the top twenty lap times above, there are three racing teams that have been the most dominant in speed over the last 3 seasons: Mercedes, McLaren, and Red Bull Racing. 

George Russell, racing for Mercedes during the 2020 season, obtained the fastest lap time of only 55 seconds. As shown in the descriptive statistics above, the average lap time for all top racer from the 2020-2022 seasons was 1.24, or 1 minute and approximately 24 seconds, with a standard deviation of about 14 seconds. Over 50% of the lap times reported from the 2020-2022 seasons fall between 1.16 and 1.33. This makes Russell's laptime over two standard deviations from the mean, quite a significant achievement in the world of Formula One racing. This can also be seen in the histogram depicting lap times above.

```python
sns.scatterplot(x="Lap Time", y="Car", data = data, hue = "Year")
plt.title("Top 20 Lap Times and Cars for 'DHL Fastest Lap' (2020-2022)")
plt.show
```

This scatterplot gives some insight into how racing team's dominance have shifted across the years...

Mercedes has been a pretty steady force to be reckoned with, appearing nine times in the top 20 ranked lap times from the 2020-2022 seasons. McLaren did well in 2020, but did not look to perform as competively in the following two seasons. Red Bull Racing seemed to emerge as a fierce competitor in the 2021-2022 seasons, appearing five times in the top 20 ranked lap times.
